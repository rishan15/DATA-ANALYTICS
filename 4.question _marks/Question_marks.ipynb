**grades are created for 'class_marks.csv' file**

```python
import pandas as pd

file_path = "class_marks.csv"  
df = pd.read_csv(file_path)
df.fillna(0, inplace=True)


max_marks = 50


df['Percentage'] = (df['Total'] / max_marks) * 50


df['Percentage'] = df['Percentage'].round(2)

def assign_grade(percentage):
    if percentage >= 40:
        return 'A'
    elif percentage >= 30:
        return 'B'
    elif percentage >= 20:
        return 'C'
    elif percentage >= 100:
        return 'D'
    else:
        return 'F'


df['Grade'] = df['Percentage'].apply(assign_grade)


print(df)
```
